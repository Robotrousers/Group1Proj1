### Project1_MoMovies - Mo Money Movies
Used Excel to extract the year from the 'release_date" column and create a new column called 'year'. Then sorted and removed all #VALUE! and 0's to clean the data. Also removed all movies older than 1931.

```python
# dependencies
import pandas as pd
from pathlib import Path
from datetime import datetime
```

```python
#import csv
tmdb_data_csv = Path("cleaned_movies_metadata.csv")
tmdb_data = pd.read_csv(tmdb_data_csv,low_memory=False)
```

```python
# check columns
print(tmdb_data.columns)
```

```python
#drop unneeded columns
tmdb_data= tmdb_data[["imdb_id", "original_title", "budget", "revenue", "year"]]
print(tmdb_data)
```

```python
#rename columns
tmdb_data = tmdb_data.rename(columns={"imdb_id": "IMDB ID", "original_title": "Title", "budget": "Budget", "revenue": "Revenue","year": "Year"})
tmdb_data.head()
```

```python
tmdb_data.dtypes
```

```python
tmdb_df = pd.DataFrame(tmdb_data)
tmdb_df
```

```python
tmdb_df.sort_values(["Year"], ascending=False)
```

```python
filtered_movies_df = tmdb_df[(tmdb_df['Year'] >= 2005) & (tmdb_df['Year'] <= 2015)]
```
